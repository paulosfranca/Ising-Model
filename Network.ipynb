```python
import matplotlib.pylab as plt
import numpy as np
import networkx as nx
```

```python
L = 3
```

```python
G1 = nx.grid_2d_graph(L, L,  periodic=True)
```

```python
G1.number_of_nodes()
```

```python
G1.nodes
```

```python
mapping=dict(zip(G1.nodes(),range(L*L)))
```

```python
mapping
```

```python
G2=nx.relabel_nodes(G1,mapping)
```

```python
G2.nodes
```

```python
G2.number_of_edges()
```

```python
G2.edges
```

```python
G2.degree()
```

```python
nx.draw(G2)
plt.show()
```

```python
G2.nodes.data()
```

```python
for i in range (L*L):
    G2.add_node(i, state=np.random.randint(0,2))
```
